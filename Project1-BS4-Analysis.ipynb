### 📊 Dataset Overview

This dataset was obtained by web scraping using the `BeautifulSoup` (BS4) and `requests` modules from  
[Box Office Mojo: The Movies Dataset](https://www.boxofficemojo.com/brand/?sort=numReleases&ref_=bo_bns__resort#table), and saved as `Movie_data.csv`.

It contains **46 rows** of data showing:
- **Brand**
- **Movie Released**
- **Total Movies Released**
- **Total Grossing for All Movies**
- **Lifetime Grossing per Movie** in **U.S. & Canada**

**Source**: Box Office Mojo

### Convert object type to Int

```python

new_data['Total_Grossing_for_all_Movies'] = [int(x.replace(",", "")) for x in new_data['Total_Grossing_for_all_Movies']]
new_data['LIfe_Time_Grossing_for_Movie'] = [int(x.replace(",", "")) for x in new_data['LIfe_Time_Grossing_for_Movie']]
new_data.info()
```

### Explore correlation through Visualisation

```python

corr_data = new_data[['Total_movie_released','Total_Grossing_for_all_Movies','LIfe_Time_Grossing_for_Movie']].corr()
corr_data
```

🔍 Interpretation
Total_movie_released and Total_Grossing_for_all_Movies have a strong positive correlation (~0.75). This suggests that brands releasing more movies tend to generate higher total gross revenue.

Total_Grossing_for_all_Movies and LIfe_Time_Grossing_for_Movie are strongly correlated (~0.80), indicating that higher lifetime gross per movie is closely linked to higher total gross revenue.

Total_movie_released and LIfe_Time_Grossing_for_Movie have a moderate positive correlation (~0.53), showing some association but weaker than the others.

```python
plt.figure(figsize=(5,5))
sns.set_context('paper',font_scale=1) 
sns.heatmap(corr_data,annot=True,cmap='plasma')
plt.title('Correlation Matrix for Movie Data Features')
plt.show()
```
